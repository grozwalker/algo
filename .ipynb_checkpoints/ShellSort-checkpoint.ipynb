```python
import math

def ShellSort(arr):
    n = len(arr)
    gap = math.floor(n / 2)
    
    
    while (gap >= 1):
    
        print('gap', gap)
        print('#########')
        for p in range(0, n):
            t = 0
            print('p', p)
            for i in range(t * gap + p, n, gap):
                print('i', i)
                tmp = arr[i]
                k = i

                while (k > 0 and arr[k - gap] > arr[k]):
                    arr[k] = arr[k - gap]
                    k = k - gap

                arr[k] = tmp
                
                print('arr', arr)
                print('--------')
                
            t = t + 1
            
        gap = math.floor(gap / 2)
    
    return arr
```

```python
ShellSort([4, 9, 5, 8, 2])
```
